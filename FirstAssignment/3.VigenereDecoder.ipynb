```python
EnglishFreqsOrder = ['e', 'a', 'r', 'i', 'o', 't', 'n', 's', 'l', 'c', 'u', 'd', 'p',
                     'm', 'h', 'g', 'b', 'f', 'y', 'w', 'k', 'v', 'x', 'z', 'j', 'q']
```

```python
def modTwentySix(x):
    if x < 0:
        return modTwentySix(x+26)
    if x >= 26:
        return modTwentySix(x-26)
    return x

def getFrequencyOrder(cipher, keySize):
    argmax = lambda l: l.index(max(l))
    freqs = list()
    letters = list()
    for i in range(keySize):
        freqs.append([])
        freqs[i] = [0] * 26
    i = 0
    for x in cipher:
        if not x.isalpha():
            continue
        freqs[i%keySize][ord(x)-ord('a')] += 1
        i += 1
    for i in range(keySize):
        letters.append([])
        for j in range(26):
            letters[i].append(chr(ord('a') + argmax(freqs[i])))
            freqs[i][argmax(freqs[i])] = 0
    return letters

def getPossibleKey(diff, keySize):
    shifts = []
    for i in range(keySize):
        shifts.append({})
        for x in diff[i]:
            if x not in list(shifts[i].keys()):
                shifts[i][x] = 1
            else:
                shifts[i][x] += 1
    key = ''
    # temp = list()
    for k in range(keySize):
        key += chr(ord('a')+modTwentySix(sorted(shifts[k].items(), key=lambda x: x[1])[-1][0]))
        # temp.append(sorted(shifts[k].items(), key=lambda x: x[1]))
    return key

def decrypt(cipher, key):
    plain = ''
    i = 0
    for x in cipher:
        if not x.isalpha():
            plain += x
        else:
            plain += chr(ord('a')+modTwentySix(ord(x)-ord(key[i%3])))
            i += 1
    return plain
```
