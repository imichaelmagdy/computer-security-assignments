```python
EnglishFreqsOrder = ['e', 'a', 'r', 'i', 'o', 't', 'n', 's', 'l', 'c', 'u', 'd', 'p',
                     'm', 'h', 'g', 'b', 'f', 'y', 'w', 'k', 'v', 'x', 'z', 'j', 'q']
```

```python
def modTwentySix(x):
    if x < 0:
        return modTwentySix(x+26)
    if x >= 26:
        return modTwentySix(x-26)
    return x

def getFrequencyOrder(cipher):
    freqs = [0] * 26
    letters = []
    argmax = lambda l: l.index(max(l))
    for x in cipher:
        if not x.isalpha():
            continue
        freqs[ord(x)-ord('a')] += 1
    for i in range(26):
        letters.append(chr(ord('a') + argmax(freqs)))
        freqs[argmax(freqs)] = 0
    return letters

def getRepeatedShifts(diff):
    shifts = {}
    for x in diff:
        if x not in list(shifts.keys()):
            shifts[x] = 1
        else:
            shifts[x] += 1
    return dict(OrderedDict(sorted(shifts.items(), key=lambda x: x[1], reverse=True)))

def decrypt(cipher, key):
    plain = ''
    for x in cipher:
        if not x.isalpha():
            plain += x
        else:
            plain += (chr(modTwentySix(ord(x)-ord('a')-key)+ord('a')))
    return plain
```
